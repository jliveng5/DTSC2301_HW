#### Question 1
In the file `https://webpages.charlotte.edu/mschuck1/classes/DTSC2301/Data/Ironman1819.csv', there is data on female finishers of the 2018 and 2019 Ironman Triathlon in Lake Placid, NY.  We will focus on the times to complete the three elements of the triathlon, swimming (Swim.Time), biking (Bike.Time) and running (Run.Time).  The units for each of these features is minutes.  The overall time (including transitions from on element to the next) is found in *Overall.Time*. Create a scatterplot for *Swim.Time* as a predictor for *Run.Time*.
```python
# import pandas, numpy and matplotlib libraries
import pandas as pd

import matplotlib.pyplot as plt
import numpy as np

from matplotlib import colors
```

```python
ironman = pd.read_csv("https://webpages.charlotte.edu/mschuck1/classes/DTSC2301/Data/Ironman1819.csv", na_values=['NA'])
# remove rows with missing data
ironman.dropna(inplace=True)
ironman.head()
```

```python
plt.scatter( x= ironman['Swim.Time'],y=ironman['Run.Time'], color="blue")

# Add labels and title
plt.xlabel('Swim time in minutes')
plt.ylabel('Run time in minutes ')
plt.title('Plot of swim time vs run time')

# Show the plot
plt.show()
```

#### Question 2
Using the Ironman data from above, fit a regression model that predicts *Overall.Time* from *Bike.Time* and add that line to a scatterplot of those two variables.  Interpret the slope and the y-intercept for these data in the context of these data.  

```python
#define response variable
y = ironman['Overall.Time']

#define predictor variables
x = ironman['Bike.Time']

# Fit linear regression via least squares with numpy.polyfit
# It returns an slope (beta_1) and intercept (beta_0)
# deg=1 means linear fit (i.e. polynomial of degree 1)
beta_1, beta_0 = np.polyfit(x, y, deg=1)

print(beta_0)
print(beta_1)
```
